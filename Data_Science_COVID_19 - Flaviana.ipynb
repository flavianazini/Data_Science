## 0.1 Fonte de dados

```python
deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
recovered = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')

display(deaths.head())
display(recovered.head())
```

### 0.2 Fonte da John Hopkins

```python
confirmed.groupby('Country/Region').sum()[last_date].head()
```

```python
deaths.groupby('Country/Region').sum()[last_date].head()
```

```python
recovered.groupby('Country/Region').sum()[last_date].head()
```

```python
#fazendo uma função que retorna os últimos dados por pais (mais recentes):

def latest_by_country(data):
  return data.groupby('Country/Region').sum()[last_date].head()

latest_by_country(confirmed).head()
```

```python
#fazendo uma função que retorna os últimos dados por pais (mais recentes),
#buscando somente a última coluna:

def latest_by_country(data):
  return data.groupby('Country/Region').sum().iloc[:,-1]

latest_by_country(confirmed).head()
```

```python
#rodando a função para os três dados - confirmed, deaths e recovered:

#criando um array
informations = [latest_by_country(confirmed), latest_by_country(deaths), latest_by_country(recovered)]

#concatenando as séries
pd.concat(informations, axis = 1)
```

```python
#renomeando os nomes das colunas:

combined = pd.concat(informations, axis = 1)
combined.columns = ['confirmed', 'deaths', 'recovered']
combined
```

```python
#para confirmar/checar que os dados estão corretos na tabela gerada acima:

display(latest_by_country(confirmed).head())
display(latest_by_country(deaths).head())
display(latest_by_country(recovered).head())
```

### 0.3 Taxa de letalidade

```python
#retirando a taxa de letalidade (%)
#ela reflete o número de casos letais dentro dos casos que foram letais e não letais

#exemplo:
# imagine que exista uma doença que em janeiro de 2020 afetou 100 pessoas
# 05 pessoas morreram, então a letalidade foi de 5%
# e como a doença continua existindo, a taxa de letalidade É de 5%

#conclusão: 
# é muito dificil falar de letalidade tentando ser preciso durante uma pandemia
# somente no começo é comum que somente os casos graves sejam reportados tornando a letalidade alta
# com o tempo a letalidade cai pq começam a investigar com exames e novos casos surgindo

```

```python
combined.sum()
```

```python
#resumo com os confirmados/mortes - como não é ideal, será chama de teste 1:

sum_up = combined.sum()
letality_rate_1 = sum_up['deaths'] / sum_up['confirmed']
print(letality_rate_1 * 100)
```

```python
#o resultado acima diz que 1,018% das pessoas do nosso conjunto de dados
#que tiveram a doença confirmada morreram
```

```python
# analisando a taxa de letalidade diminuido/aumentando ao longo do tempo
```

```python
#somando os casos por dia, e jogando fora a lat, long e a s duas ultimas linhas:

confirmed.sum()[3:-2]
```

```python
#fazendo o mesmo para os recuperados e mortos, porém eles não possum as últimas linhas:

recovered.sum()[3:], deaths.sum()[2:]
```

```python
pd.DataFrame([confirmed.sum()[3:-2],recovered.sum()[3:], deaths.sum()[2:]])
```

```python
#invertando linha x coluna, e colocando nomes nas colunas

pd.DataFrame([confirmed.sum()[3:-2],recovered.sum()[3:], deaths.sum()[2:]],
             index = ['confirmed', 'recovered', 'deaths']).T
```

```python
grouped_per_day = pd.DataFrame([confirmed.sum()[3:-2],recovered.sum()[3:], deaths.sum()[2:]],
             index = ['confirmed', 'recovered', 'deaths']).T
grouped_per_day.head()
```

```python
#calculando a letalidade do mundo inteiro:

grouped_per_day['letality_rate_1'] = grouped_per_day['deaths'] / grouped_per_day['confirmed'] * 100
grouped_per_day['letality_rate_2'] = grouped_per_day['deaths'] / (grouped_per_day['deaths'] +grouped_per_day['recovered']) * 100

grouped_per_day.head()
```

```python
#criando gráfico para letalidade 1:

grouped_per_day['letality_rate_1'].plot()
plt.title('Letalidade pessimista e orimista no mundo todo')
plt.show()
```

```python
grouped_per_day['letality_rate_1'].plot()
grouped_per_day['letality_rate_2'].plot()
plt.title('Letalidade pessimista e orimista no mundo todo')
plt.show()
```

```python
plt.figure(figsize=(10,5))
grouped_per_day['letality_rate_1'].plot()
grouped_per_day['letality_rate_2'].plot()
plt.legend(['Mortes / Casos confirmados', 'Mortes / Mortes + Recuperados'] )
plt.title('Letalidade pessimista e orimista no mundo todo')
plt.show()
```

```python
plt.figure(figsize=(10,5))
grouped_per_day['letality_rate_1'].plot()
grouped_per_day['letality_rate_2'].plot()
plt.legend(['Mortes / Casos confirmados', 'Mortes / Mortes + Recuperados'] )
plt.title('Letalidade pessimista e orimista no mundo todo')
plt.grid(linestyle = '-')
plt.show()
```
